```python
import pandas as pd
```

```python
df = pd.read_excel("KERALA/KERALA - 2021.xlsx")
df_clean = df[df["STATE/UT NAME"]!="TURN OUT"].copy()

df_clean["AC TOTAL"] = df_clean.groupby("AC NO.")["TOTAL"].transform("sum")
df_clean["VOTE SHARE"] = df_clean["TOTAL"]/df_clean["AC TOTAL"]

df_clean['is_winner'] = (df_clean['TOTAL'] == df_clean.groupby('AC NO.')['TOTAL'].transform('max')).astype(int)
df_clean.columns

df_clean.drop(["STATE/UT NAME", "SYMBOL", "GENERAL", "POSTAL", "% VOTES POLLED"], axis=1, inplace=True)
df_clean
```

```python
df_2016 = pd.read_excel("KERALA/KERALA - 2016.xlsx")

rename_map_2016 = {
    'Constituency No.': 'AC NO.',
    'Constituency Name': 'AC NAME',
    'Candidate Name': 'CANDIDATE NAME',
    'Candidate Sex': 'SEX',
    'Candidate Age': 'AGE',
    'Candidate Category': 'CATEGORY',
    ' Party Name': 'PARTY',
    ' Total Valid Votes': 'TOTAL',
    'Total Electors': 'TOTAL ELECTORS'
}

df_2016 = df_2016.rename(columns=rename_map_2016)
df_2016.columns
df_2016.drop([" VALID VOTES POLLED in Postal", " VALID VOTES POLLED in General", "Total Votes"], axis=1, inplace=True)
df_2016.columns

df_2016["AC TOTAL"] = df_2016.groupby("AC NO.")["TOTAL"].transform("sum")
df_2016["VOTE SHARE"] = df_2016["TOTAL"]/df_2016["AC TOTAL"]

df_2016['is_winner'] = (df_2016['TOTAL'] == df_2016.groupby('AC NO.')['TOTAL'].transform('max')).astype(int)
df_2016.columns
```
